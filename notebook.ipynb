```python
stock = {
"pains au chocolat": 500,
"croissants": 400,
"baguettes": 500,
"beignets": 200
}
```

```python
def afficher_inventaire(stock):
    print("Inventaire de la boulangerie :")
    for produit, quantite in stock.items():
        print(f"{produit} : {quantite}")
afficher_inventaire(stock)
```

```python
def fabriquer(produit, quantite, stock):
    if produit in stock:
        stock[produit] += quantite

fabriquer("croissants", 50, stock)
```

```python
def vendre(produit, quantite, prix):
    total = quantite * prix

    # Réduction si quantité >= 10
    if quantite >= 10:
        total = total * 0.25

    print(f"Vente de {quantite} {produit} pour un total de {total} €")
    return total

vendre("beignets", 20, 2.10)
```

```python
tresorerie = 0 

def vendre(produit, quantite, prix, tresorerie):
    total = quantite * prix

    if quantite < 10:
        total = total * 0.25

    tresorerie += total
    print(f"Vente de {quantite} {produit} pour un total de {total} €")
    print(f"Trésorerie actuelle : {tresorerie} €")

    return tresorerie
tresorerie = vendre("beignets", 70, 2.10, tresorerie)
tresorerie = vendre("donuts", 7, 3, tresorerie)
tresorerie = vendre("croissants",100, 2.10, tresorerie)
```
