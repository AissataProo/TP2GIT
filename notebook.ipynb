```python
stock = {
"pains au chocolat": 500,
"croissants": 400,
"baguettes": 500,
"beignets": 200
}
```

```python
def afficher_inventaire(stock):
    print("Inventaire de la boulangerie :")
    for produit, quantite in stock.items():
        print(f"{produit} : {quantite}")
afficher_inventaire(stock)
```

```python
def fabriquer(produit, quantite, stock):
    if produit in stock:
        stock[produit] += quantite

fabriquer("croissants", 50, 1)
```
